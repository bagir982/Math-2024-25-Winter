### **1. Compute the derivatives**

1. $y(x) = -3x + 3$  
   $y'(x) = -3$

2. $y(x) = \pi x + \sin(1)$  
   $y'(x) = \pi$

3. $y(x) = 4 + \sin(2)$  
   $y'(x) = 0$ (since $4 + \sin(2)$ is a constant)

4. $y(x) = 2x^3 - 3x^2 + 8x - 9$  
   $y'(x) = 6x^2 - 6x + 8$

5. $y(x) = 6x^{1/3}$  
   $y'(x) = 2x^{-2/3}$

6. $y(x) = \sqrt{x}$  
   $y'(x) = \frac{1}{2\sqrt{x}}$

7. $y(x) = \cos(x) + \sin(x)$  
   $y'(x) = -\sin(x) + \cos(x)$

8. $y(x) = 2\sin(x)\cos(x)$  
   $y'(x) = 2\cos(2x)$ (using the double angle formula)

9. $y(x) = x\sin(x)$  
   $y'(x) = \sin(x) + x\cos(x)$

10. $y(x) = (x + 1)^2$  
    $y'(x) = 2(x + 1)$

11. $y(x) = \frac{x}{x + 1}$  
    $y'(x) = \frac{1}{(x + 1)^2}$

12. $y(x) = (x + 1)\exp(x)$  
    $y'(x) = (x + 1)\exp(x) + \exp(x) = (x + 2)\exp(x)$

13. $y(x) = \sin(x^2)$  
    $y'(x) = 2x\cos(x^2)$

14. $y(x) = \exp(-2x)$  
    $y'(x) = -2\exp(-2x)$

15. $y(x) = \frac{1}{\sin(x + 1)}$  
    $y'(x) = -\frac{\cos(x + 1)}{\sin^2(x + 1)}$

16. $y(x) = \sqrt{2x + 1}$  
    $y'(x) = \frac{1}{\sqrt{2x + 1}}$

### **2. Prove:**  
$\frac{d}{dx} (\ln(\sin(x))) = \cot(x)$

- **Solution:**  
  Apply the chain rule:

  $\frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)$

### **3. Verify:**  
For $f(x) = \cos(x)$, check if $f''(x) = -f(x)$.

- **Solution:**  
  $f'(x) = -\sin(x)$  
  $f''(x) = -\cos(x) = -f(x)$

### **4. Limits using L'Hospital's Rule**

1. $\displaystyle \lim_{x \to 0} \frac{\sin(x)}{x}$  
   Apply L'Hospital's Rule:  
   $\lim_{x \to 0} \frac{\sin(x)}{x} = 1$

2. $\displaystyle \lim_{x \to \infty} \frac{\ln(x)}{x}$  
   Apply L'Hospital's Rule:  
   $\lim_{x \to \infty} \frac{\ln(x)}{x} = 0$

3. $\displaystyle \lim_{x \to \infty} \frac{\exp(x)}{x}$  
   Apply L'Hospital's Rule:  
   $\lim_{x \to \infty} \frac{\exp(x)}{x} = \infty$

### **5. Velocity and acceleration of a particle**

Given position function: $x(t) = 3t^2 - 6t + 1$

- **Velocity:**  
  $V(t) = \frac{dx(t)}{dt} = 6t - 6$

- **Acceleration:**  
  $a(t) = \frac{dV(t)}{dt} = 6$

- **At $t = 2$:**
  - Velocity: $V(2) = 6(2) - 6 = 6$
  - Acceleration: $a(2) = 6$
