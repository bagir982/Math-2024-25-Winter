### **1. Equation of the plane passing through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$**

- **Explanation:**  
  The normal vector to the plane is:

  $ \mathbf{n} = (B - A) \times (C - A) = [4, 0, -4] $

  The equation of the plane is:

  $ 4x - 4z + 8 = 0 $

### **4. Line of intersection of planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$**

- **Explanation:**  
  Solve the system of equations:

  $ \begin{cases} 2x + 3y + 4z = 5 \\ 3x + 4y + 2z = 6 \end{cases} $

  The line of intersection is parametrized by:

  $ y = \frac{7}{5} - \frac{4}{5}x, \quad z = \frac{x}{10} + \frac{1}{5} $

### **5. Equation of the plane passing through point $A(1, 2, 3)$ and parallel to vectors $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$**

- **Explanation:**  
  The normal vector to the plane is:

  $ \mathbf{n} = \vec{v_1} \times \vec{v_2} = [-1, 1, 1] $

  The equation of the plane is:

  $ -x + y + z - 4 = 0 $

### **7. Distance from point $A(1, 2, 3)$ to the plane $2x + 3y + 4z = 5$**

- **Explanation:**  
  The formula for the distance from a point to a plane $Ax + By + Cz + D = 0$ is:

  $ d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}} $

  Substituting values:

  $ d = \frac{15}{\sqrt{29}} \approx 0.88 $

### **9. Angle between planes $x + y + z = 1$ and $x = 0$**

- **Explanation:**  
  The formula for the angle between two planes is:

  $ \theta = \cos^{-1} \left( \frac{|\mathbf{n_1} \cdot \mathbf{n_2}|}{\|\mathbf{n_1}\| \|\mathbf{n_2}\|} \right) $

  Substituting values, the angle is approximately:

  $ \theta = 54.74^\circ $
