### **1. Equation of the line passing through points $A(1, 2)$ and $B(3, 4)$**

- **Explanation:**  
  The slope of the line is:

  $ m = \frac{4 - 2}{3 - 1} = 1 $

  The equation of the line is:

  $ y = x + 1 $

---

### **2. Equation of the line passing through point $A(1, 2)$ and parallel to the line $y = 2x + 3$**

- **Explanation:**  
  The slope of the line is the same as the given line, $m = 2$.

  The equation of the line is:

  $ y = 2(x - 1) + 2 = 2x $

---

### **3. Equation of the line passing through point $A(1, 2)$ and perpendicular to the line $y = 2x + 3$**

- **Explanation:**  
  The slope of the perpendicular line is the negative reciprocal of $2$, which is $-1/2$.

  The equation of the line is:

  $ y = -\frac{1}{2}(x - 1) + 2 = -\frac{1}{2}x + 2.5 $

---

### **4. Find the intersection point and angle between the lines $y = 2x + 3$ and $y = 3x + 2$**

- **Intersection Point:**  
  Solve the system of equations:

  $ \begin{cases} y = 2x + 3 \\ y = 3x + 2 \end{cases} $

  The intersection point is $ (1, 5) $.

- **Angle Between Lines:**  
  The formula for the angle between two lines with slopes $m_1$ and $m_2$ is:

  $ \theta = \tan^{-1} \left( \frac{|m_2 - m_1|}{1 + m_1 m_2} \right) $

  Substituting $m_1 = 2$ and $m_2 = 3$, the angle is approximately $25.54^\circ$.

---

### **5. Equation of the line passing through point $A(1, 2)$ and parallel to vector $\mathbf{v} = [2, 3]$**

- **Explanation:**  
  The slope of the line is:

  $ m = \frac{3}{2} $

  The equation of the line is:

  $ y = \frac{3}{2}x + \frac{1}{2} $

### **7. Distance from point $A(1, 2)$ to the line $y = 2x + 3$**

- **Explanation:**  
  The formula for the distance from a point $(x_1, y_1)$ to a line $Ax + By + C = 0$ is:

  $ d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}} $

  For point $A(1, 2)$ and line $y = 2x + 3$, the distance is:

  $ d = \frac{3\sqrt{5}}{5} $

---

### **8. Equation of the line passing through points $A(2, 0)$ and $B(0, 3)$**

- **Explanation:**  
  The slope of the line is:

  $ m = \frac{3 - 0}{0 - 2} = -\frac{3}{2} $

  The equation of the line is:

  $ y = -\frac{3}{2}x + 3 $

---

### **9. Calculate the angle between the line $y = x + 3$ and the $Ox$ axis**

- **Explanation:**  
  The slope of the line is $1$. The angle with the $Ox$ axis is:

  $ \theta = \tan^{-1}(1) = 45^\circ $
