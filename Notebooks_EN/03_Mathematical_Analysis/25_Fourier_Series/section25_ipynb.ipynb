### **1. Solve the first-order ordinary differential equations**

1. $ \frac{dy}{dx} = y(x) $  
   - **Explanation:** This is a simple first-order differential equation. We apply the exponential growth rule, which gives the solution:  
     $ y(x) = C_1 e^x $

2. $ \frac{dy}{dx} = \frac{1}{2y(x)} $  
   - **Explanation:** Multiply both sides by $ 2y(x) $ to separate the variables:  
     $ 2y(x) \, dy = dx $  
     Integrate both sides:  
     $ y(x) = \pm \sqrt{C_1 + x} $

### **2. Solve the first-order ODEs using separation of variables**

1. $ \frac{dy}{dx} = \frac{x}{y} $  
   - **Explanation:** Separate the variables:  
     $ y \, dy = x \, dx $  
     Integrate both sides:  
     $ \int y \, dy = \int x \, dx $  
     $ \frac{y^2}{2} = \frac{x^2}{2} + C_1 $  
     The solution is: $ y(x) = \pm \sqrt{C_1 + x^2} $

2. $ \frac{dy}{dx} = \frac{y}{x} $  
   - **Explanation:** Separate the variables:  
     $ \frac{dy}{y} = \frac{dx}{x} $  
     Integrate both sides:  
     $ \ln |y| = \ln |x| + C_1 $  
     The solution is: $ y(x) = C_1 x $

3. $ \frac{dy}{dx} = xy $  
   - **Explanation:** Separate the variables:  
     $ \frac{dy}{y} = x \, dx $  
     Integrate both sides:  
     $ \ln |y| = \frac{x^2}{2} + C_1 $  
     The solution is: $ y(x) = C_1 e^{\frac{x^2}{2}} $

### **3. Solve the second-order ordinary differential equations**

1. $ y''(x) + y'(x) = 0 $  
   - **Explanation:** This is a second-order linear differential equation. We integrate it twice:  
     First integration: $ y'(x) = C_2 e^{-x} $  
     Second integration: $ y(x) = C_1 + C_2 e^{-x} $

2. $ y''(x) - y(x) = 0 $  
   - **Explanation:** The solution involves exponential functions:  
     $ y(x) = C_1 e^{-x} + C_2 e^{x} $

3. $ y''(x) + \omega^2 y(x) = 0 $  
   - **Explanation:** The solution involves complex exponentials or trigonometric functions:  
     $ y(x) = C_1 e^{-i\omega x} + C_2 e^{i\omega x} $  
     Alternatively, it can be expressed as:  
     $ y(x) = A \cos(\omega x) + B \sin(\omega x) $

### **4. Verify that $ \psi(t, x) = A \cos(\omega t + kx) $ satisfies the wave equation**

Wave equation:  
$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0 $

- Substitute $ \psi(t, x) = A \cos(\omega t + kx) $ into the equation.
- The second derivative with respect to $ t $ is:  
  $ \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx) $

- The second derivative with respect to $ x $ is:  
  $ \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx) $

- The equation becomes:  
  $ -A \omega^2 \cos(\omega t + kx) + v^2 A k^2 \cos(\omega t + kx) = 0 $

- For the equation to hold, $ v^2 = \frac{\omega^2}{k^2} $, confirming that $ \psi(t, x) $ is a solution.
