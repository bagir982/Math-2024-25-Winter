## Tasks Related to Equations of Lines on a Plane

### 1. The line passes through points $A(1, 2)$ and $B(3, 4)$. Find the equation of the line.

The slope of the line passing through points $A(x_1, y_1)$ and $B(x_2, y_2)$ is:
$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{4 - 2}{3 - 1} = 1
$$

Using the point-slope form:
$$
y - y_1 = m(x - x_1)
$$

Substitute $m = 1$, $x_1 = 1$, and $y_1 = 2$:
$$
y - 2 = 1(x - 1) \implies y = x + 1
$$

**Answer**: The equation of the line is $y = x + 1$.

---

### 2. The line passes through point $A(1, 2)$ and is parallel to the line $y = 2x + 3$. Find the equation of the line.

Parallel lines have the same slope. Since $y = 2x + 3$ has slope $m = 2$, the equation of the line is:
$$
y - y_1 = m(x - x_1)
$$

Substitute $m = 2$, $x_1 = 1$, and $y_1 = 2$:
$$
y - 2 = 2(x - 1) \implies y = 2x
$$

**Answer**: The equation of the line is $y = 2x$.

---

### 3. The line passes through point $A(1, 2)$ and is perpendicular to the line $y = 2x + 3$. Find the equation of the line.

The slope of a line perpendicular to $y = 2x + 3$ is the negative reciprocal of $2$:
$$
m = -\frac{1}{2}
$$

Using the point-slope form:
$$
y - y_1 = m(x - x_1)
$$

Substitute $m = -\frac{1}{2}$, $x_1 = 1$, and $y_1 = 2$:
$$
y - 2 = -\frac{1}{2}(x - 1) \implies y = -\frac{1}{2}x + \frac{5}{2}
$$

**Answer**: The equation of the line is $y = -\frac{1}{2}x + \frac{5}{2}$.

---

### 4. Intersection of lines $y = 2x + 3$ and $y = 3x + 2$, and the angle between them.

**Intersection Point**:  
Equate the two equations:
$$
2x + 3 = 3x + 2 \implies x = 1, \, y = 5
$$

The intersection point is $(1, 5)$.

**Angle Between the Lines**:  
The angle between two lines is given by:
$$
\tan \theta = \left| \frac{m_2 - m_1}{1 + m_1 m_2} \right|
$$

Here, $m_1 = 2$ and $m_2 = 3$:
$$
\tan \theta = \left| \frac{3 - 2}{1 + 2 \cdot 3} \right| = \frac{1}{7}
$$

The angle $\theta$ is:
$$
\theta = \tan^{-1}\left(\frac{1}{7}\right)
$$

### 5. Write the equation of the line passing through point $A(1, 2)$ and parallel to the vector $\vec{v} = [2, 3]$.

A vector parallel to the line gives its direction. The direction vector $\vec{v} = [2, 3]$ corresponds to slope:
$$
m = \frac{3}{2}
$$

Using the point-slope form:
$$
y - y_1 = m(x - x_1)
$$

Substitute $m = \frac{3}{2}$, $x_1 = 1$, and $y_1 = 2$:
$$
y - 2 = \frac{3}{2}(x - 1) \implies y = \frac{3}{2}x + \frac{1}{2}
$$

**Answer**: The equation of the line is $y = \frac{3}{2}x + \frac{1}{2}$.

---

### 6. We have the line $y = 2x + 3$. Find an example of a line perpendicular and parallel to it.

**Parallel Line**:  
A parallel line has the same slope $m = 2$. For example:
$$
y = 2x + 1
$$

**Perpendicular Line**:  
A perpendicular line has slope $m = -\frac{1}{2}$. For example:
$$
y = -\frac{1}{2}x + 1
$$

---

### 7. Distance from point $A(1, 2)$ to the line $y = 2x + 3$.

The distance from a point $(x_1, y_1)$ to a line $ax + by + c = 0$ is given by:
$$
d = \frac{|ax_1 + by_1 + c|}{\sqrt{a^2 + b^2}}
$$

Rewriting $y = 2x + 3$ as:
$$
2x - y + 3 = 0
$$

Substitute $a = 2$, $b = -1$, $c = 3$, $x_1 = 1$, $y_1 = 2$:
$$
d = \frac{|2(1) - (2) + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|3|}{\sqrt{5}} = \frac{3}{\sqrt{5}}
$$

---

### 8. The line intersects the coordinate axes at points $A(2, 0)$ and $B(0, 3)$. Find the equation of the line.

The slope of the line passing through $A(2, 0)$ and $B(0, 3)$ is:
$$
m = \frac{3 - 0}{0 - 2} = -\frac{3}{2}
$$

Using the point-slope form:
$$
y - y_1 = m(x - x_1)
$$

Substitute $m = -\frac{3}{2}$, $x_1 = 2$, and $y_1 = 0$:
$$
y - 0 = -\frac{3}{2}(x - 2) \implies y = -\frac{3}{2}x + 3
$$

**Answer**: The equation of the line is $y = -\frac{3}{2}x + 3$.
